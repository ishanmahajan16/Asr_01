# Training MFCC Features

## Importing Libraries

```python
from sklearn.mixture import GaussianMixture
import pandas as pd
import numpy as np
import math
import pickle
```

## Loading Features and Labels

```python

def transform(file1,file2):
    x=np.load(file1).tolist()
    y=np.load(file2).tolist()
    return pd.DataFrame({"feature":x,"label":y})


pathf=["features/mfcc/features.npy","features/mfcc_delta/features_delta.npy","features/mfcc_delta_delta/features_delta_delta.npy"]
pathl=["features/mfcc/labels.npy","features/mfcc_delta/labels_delta.npy","features/mfcc_delta_delta/labels_delta_delta.npy"]
df = [transform(pathf[0],pathl[0]),transform(pathf[1],pathl[1]),transform(pathf[2],pathl[2])]

```

## Training and Dumping Models

```python

delta_with=["./models/mfcc/with_energy_coeff/","./models/mfcc_delta/with_energy_coeff/","./models/mfcc_delta_delta/with_energy_coeff/"]
delta_without=["./models/mfcc/without_energy_coeff/","./models/mfcc_delta/without_energy_coeff/","./models/mfcc_delta_delta/without_energy_coeff/"]
coeff_remove=[[0],[0,14],[0,14,27]]



for j in range(0,3):
    
    print(j)
    
    
    phoneme=df[j]['label'].unique()
    phoneme.sort()
    
    for ph in phoneme:
        
        print(ph)
        
        temp=df[j][df[j]['label']==ph]
        features=np.array(temp["feature"].tolist())
        labels = np.array(temp["label"].tolist())
        
        models=GaussianMixture(n_components=64,covariance_type='diag',max_iter=100)
        models.fit(features)
        pickle.dump(models, open(delta_with[j]+'/64/'+ph+".pkl", 'wb'))
        

        features=np.delete(features,coeff_remove[j], axis=1)
        
        models=GaussianMixture(n_components=64,covariance_type='diag',max_iter=100)
        models.fit(features)
        pickle.dump(models, open(delta_without[j]+'/64/'+ph+".pkl", 'wb'))
       
        
        if j==0:
            
            comp=[2,4,8,16,32,128,256]
            
            for i in comp:
                
                print(i)
           
                models=GaussianMixture(n_components=i,covariance_type='diag',max_iter=100)
                models.fit(features)
                pickle.dump(models, open(delta_without[j]+'/'+(str)(i)+"/"+ph+".pkl", 'wb'))
                
        
        
        
```
